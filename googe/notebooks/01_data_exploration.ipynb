# 01 - Data Exploration

This notebook explores the sample claims dataset and visualizes patterns.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Load sample claims
with open('../data/sample_claims.json', 'r') as f:
    data = json.load(f)

claims = data['claims']
print(f"Loaded {len(claims)} claims")
```

```python
# Convert to DataFrame
df = pd.DataFrame(claims)
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.head()
```

```python
# Verdict distribution
verdicts = Counter(df['initial_verdict'])
plt.figure(figsize=(8, 4))
plt.bar(verdicts.keys(), verdicts.values())
plt.title('Verdict Distribution')
plt.xlabel('Verdict')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Platform distribution
platforms = Counter(df['source_platform'])
plt.figure(figsize=(8, 4))
plt.pie(platforms.values(), labels=platforms.keys(), autopct='%1.1f%%')
plt.title('Source Platform Distribution')
plt.tight_layout()
plt.show()
```
