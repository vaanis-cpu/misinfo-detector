```python
import sys
sys.path.insert(0, '..')

import torch
import numpy as np
from src.graph import ClaimGraph
from src.types import create_claim, ClaimNode, PropagationEdge, GraphSnapshot
from datetime import datetime
```

```python
# Create sample graph for training demo
graph = ClaimGraph()

# Add root claim
root = create_claim(
    content="Sample claim for training",
    source_url="https://twitter.com/test/1",
    source_platform="twitter",
    author_id="user1"
)
graph.add_claim(root)

# Generate propagation tree
graph.generate_propagation_tree(root.claim_id, num_nodes=50)

print(f"Graph: {graph.graph.number_of_nodes()} nodes, {graph.graph.number_of_edges()} edges")
```

```python
# Get snapshot for training
snapshot = graph.get_snapshot(root.claim_id, depth=5)
print(f"Snapshot: {snapshot.node_count} nodes, {snapshot.edge_count} edges")
```
